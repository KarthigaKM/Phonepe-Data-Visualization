AGGREGATED TRANSACTIONS

```python
data_aggregate_trans = "C:/Users/ADMIN/phonepeplus/data/aggregated/transaction/country/india/state/"

column_names_1 = {'State': [], 'Year': [], 'Quarter': [], 'Transaction_type': [], 'Transaction_count': [],
            'Transaction_amount': []}

# Traverse the state directories within data_aggregate_transaction directories
for state in os.listdir(data_aggregate_trans):
    state_dir = os.path.join(data_aggregate_trans, state)
    if os.path.isdir(state_dir):
        # Traverse the year directories within each state
        for year in os.listdir(state_dir):
            year_dir = os.path.join(state_dir, year)
            if os.path.isdir(year_dir):
                # Traverse the json file within each year
                for file in os.listdir(year_dir):
                    if file.endswith(".json"):
                                json_file = os.path.join(year_dir, file)
                                with open(json_file, "r") as data:
                                    json_data1 = json.load(data)
                                    try:
                                         
                                        for i in json_data1 ['data']['transactionData']:
                                            name = i['name']
                                            count = i['paymentInstruments'][0]['count']
                                            amount = i['paymentInstruments'][0]['amount']
                                            amount = "{:.2f}".format(amount)
                                            column_names_1 ['Transaction_type'].append(name)
                                            column_names_1 ['Transaction_count'].append(count)
                                            column_names_1 ['Transaction_amount'].append(amount)
                                            column_names_1 ['State'].append(state)
                                            column_names_1 ['Year'].append(year)
                                            column_names_1 ['Quarter'].append(int(file.strip('.json')))
                                    except:
                                        pass
df_aggregate_trans = pd.DataFrame(column_names_1 )
```
