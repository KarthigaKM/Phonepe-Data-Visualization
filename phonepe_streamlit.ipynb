```python
%%writefile phonepe_main.py
import pandas as pd
import csv
import numpy as np
import mysql.connector 
from mysql.connector import (connection)
import streamlit as st
from streamlit_option_menu import option_menu
from PIL import Image
import plotly.express as px
import os
import json
from PIL import Image

   
#Create page configuration
icon= Image.open('C:/Users/ADMIN/images/phonepelogo.png') 

st.set_page_config(
                   page_title="PhonePe Pulse",
                   page_icon=icon,
                   layout="wide",
                   initial_sidebar_state="expanded",
                   )
st.title(":violet[PhonePe Pulse | THE BEAT OF PROGRESS]")


# OPTION MENUS
with st.sidebar:
    selected = option_menu("Menu", ["Home","Chart_Visualizations","Map_Visualizations"],
                          default_index=0,
                          orientation="vertical",
                          styles={"nav-link": {"font-size": "30px", "text-align": "centre", "margin": "0px", 
                                                "--hover-color": "#800080"},
                                   "icon": {"font-size": "40px"},
                                   "container" : {"max-width": "4000px"},
                                   "nav-link-selected": {"background-color": "#800080"}})

#Connect to MySQL
conn = connection.MySQLConnection(host="localhost", user="root", port='3306', database="YOUR DATABSE NAME", password="YOUR PASSWORD")
cur = conn.cursor()
        

import streamlit as st
import plotly.express as px

#india_states = json.load(open("C:/Users/ADMIN/india map/states_india.geojson",'r'))
with open("C:/Users/ADMIN/india map/states_india.geojson", "r") as f:
    geojson_data = json.load(f)

state_map_id={}
for feature in geojson_data["features"]:
    feature["id"]=feature["properties"]["state_code"]
    state_map_id[feature["properties"]["st_nm"]]=feature["id"]
    
state_names = list(state_map_id.keys())
print(state_names)
# Specify the output file path
output_file = 'state_names.csv'

# Write the state names to the CSV file
with open(output_file, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['State Names'])  # Write the header
    writer.writerows([[state] for state in state_names])  # Write the state names
    
    
    
    
if selected == "Home":
    # Title Image
    image = Image.open('C:/Users/ADMIN/images/Create-PhonePe-Account.png')  # Replace 'path/to/your/image.jpg' with the actual file path of your image
    st.image(image, caption=' Phonepe ', use_column_width=True)

    st.markdown(":blue[Project Title] : Phonepe Pulse Data Visualization and Exploration")
    st.markdown(":blue[Domain] : Fintech")
    st.markdown(":blue[Technologies ] : Github Cloning, Python, Pandas, MySQL,mysql-connector-python, Streamlit, and Plotly.")
    st.markdown(":blue[Overview] : Use the Streamlit and Plotly libraries in Python to create an interactive and visually appealing dashboard. Plotly's built-in geo map functions can be used to display the data on a map and Streamlit can be usedto create a user-friendly interface with multiple dropdown options for users toselect different facts and figures to display.")

if selected == "Chart_Visualizations":
     #Create a dropdown menu with visualization options
    tab1,tab2,tab3,tab4,tab5,tab6= st.tabs(['Transaction Type Chart',
                                'User by Brand Chart',
                                'Top Transactions by District',
                                'Top Users by District',
                                'Top Transactions by Pincode',
                                'Top Users by  Pincode'])

    with tab1:
            selected_state1 = st.selectbox("STATES1",
                             ('andaman-&-nicobar-islands','andhra-pradesh','arunachal-pradesh','assam','bihar',
                              'chandigarh','chhattisgarh','dadra-&-nagar-haveli-&-daman-&-diu','delhi','goa','gujarat','haryana',
                              'himachal-pradesh','jammu-&-kashmir','jharkhand','karnataka','kerala','ladakh','lakshadweep',
                              'madhya-pradesh','maharashtra','manipur','meghalaya','mizoram',
                              'nagaland','odisha','puducherry','punjab','rajasthan','sikkim',
                              'tamil-nadu','telangana','tripura','uttar-pradesh','uttarakhand','west-bengal'),index=30)
            
        
        
            query = f"SELECT Transaction_type, sum(Transaction_count) as Total_Transactions, sum(Transaction_amount) as Total_amount from aggregated_trans WHERE Year BETWEEN 2018 AND 2022 AND Quarter BETWEEN 1 AND 4 AND State = '{selected_state1}' GROUP by State, Year,Quarter,Transaction_type ORDER by State,Transaction_type"
            cur.execute(query)
            results = cur.fetchall()
            df_results = pd.DataFrame(results, columns=['Transaction_type', 'Total_Transactions','Total_amount'])
            #df_results['var_amount'] = np.log10(df_results['Total_amount'])
            fig1 = px.pie(df_results, title=selected_state1, names='Transaction_type',hover_data='Total_amount', values='Total_Transactions')
            col1, col2=st.columns(2)
            col1.plotly_chart(fig1,use_container_width=True)
            col2.markdown('**Aggregated Transactions by Transaction Type**')
            col2.write(df_results)
            
            
    with tab2: 
            selected_state2 = st.selectbox("STATES2",
                             ('andaman-&-nicobar-islands','andhra-pradesh','arunachal-pradesh','assam','bihar',
                              'chandigarh','chhattisgarh','dadra-&-nagar-haveli-&-daman-&-diu','delhi','goa','gujarat','haryana',
                              'himachal-pradesh','jammu-&-kashmir','jharkhand','karnataka','kerala','ladakh','lakshadweep',
                              'madhya-pradesh','maharashtra','manipur','meghalaya','mizoram',
                              'nagaland','odisha','puducherry','punjab','rajasthan','sikkim',
                              'tamil-nadu','telangana','tripura','uttar-pradesh','uttarakhand','west-bengal'),index=30)
            
        
            
            query= f"SELECT Brands,sum(Count) AS Total_Users,avg(Percentage)*100 as Average_Percentage FROM aggregated_user WHERE Year BETWEEN 2018 AND 2022 AND Quarter BETWEEN 1 AND 4 AND State ='{selected_state2}'GROUP by State, Year, Quarter,Brands ORDER BY State, Brands"
            cur.execute(query)
            results = cur.fetchall()
            df_results = pd.DataFrame(results, columns=['Brands','Total_Users','Average_Percentage'])
            fig2 = px.bar(df_results, x='Total_Users', y='Average_Percentage', color='Brands',
                         title='selected_state2')
            col1, col2=st.columns(2)
            col1.plotly_chart(fig2,use_container_width=True)
            col2.markdown('**Aggregated User Data Brand**')
            col2.write(df_results)
            
    with tab3:  
            selected_state3 = st.selectbox("STATES3",
                             ('andaman-&-nicobar-islands','andhra-pradesh','arunachal-pradesh','assam','bihar',
                              'chandigarh','chhattisgarh','dadra-&-nagar-haveli-&-daman-&-diu','delhi','goa','gujarat','haryana',
                              'himachal-pradesh','jammu-&-kashmir','jharkhand','karnataka','kerala','ladakh','lakshadweep',
                              'madhya-pradesh','maharashtra','manipur','meghalaya','mizoram',
                              'nagaland','odisha','puducherry','punjab','rajasthan','sikkim',
                              'tamil-nadu','telangana','tripura','uttar-pradesh','uttarakhand','west-bengal'),index=30)
            
            query = f"SELECT District, sum(Transaction_count) as Total_Transactions, sum(Transaction_amount) as Total_amount from top_trans WHERE Year BETWEEN 2018 AND 2022 AND Quarter BETWEEN 1 AND 4  AND State = '{selected_state3}' GROUP by State, District,Year,Quarter ORDER by State,District"
            cur.execute(query)
            results = cur.fetchall()
            df_results = pd.DataFrame(results, columns=['District', 'Total_Transactions','Total_amount'])
            df_results['var_amount'] = np.log10(df_results['Total_amount'])
            fig3 = px.pie(df_results, values='var_amount', names='District', hover_data='Total_Transactions',
                  title=  selected_state3)
            col1, col2=st.columns(2)
            col1.plotly_chart(fig3,use_container_width=True)
            col2.markdown('**Top Transactions by District**')
            col2.write(df_results)
            
            
    with tab4 :      
            selected_state4 = st.selectbox("STATES4",
                             ('andaman-&-nicobar-islands','andhra-pradesh','arunachal-pradesh','assam','bihar',
                              'chandigarh','chhattisgarh','dadra-&-nagar-haveli-&-daman-&-diu','delhi','goa','gujarat','haryana',
                              'himachal-pradesh','jammu-&-kashmir','jharkhand','karnataka','kerala','ladakh','lakshadweep',
                              'madhya-pradesh','maharashtra','manipur','meghalaya','mizoram',
                              'nagaland','odisha','puducherry','punjab','rajasthan','sikkim',
                              'tamil-nadu','telangana','tripura','uttar-pradesh','uttarakhand','west-bengal'),index=30)
            
            query = f"SELECT District, sum(RegisteredUsers) as Total_Users FROM  top_user WHERE Year BETWEEN 2018 AND 2022 AND Quarter BETWEEN 1 AND 4  AND State = '{selected_state4}' GROUP by State, District,Year,Quarter order by State,District"
            cur.execute(query)
            results = cur.fetchall()
            df_results = pd.DataFrame(results, columns=['District', 'Total_Users '])
            
            fig4 = px.pie(df_results, values='Total_Users ', names='District', hover_data='Total_Users ',
                  title=  selected_state4)
            col1, col2=st.columns(2)
            col1.plotly_chart(fig4,use_container_width=True)
            col2.markdown('**Top Users by District**')
            col2.write(df_results)
            
     
    with tab5:
            selected_state5 = st.selectbox("STATES5",
                             ('andaman-&-nicobar-islands','andhra-pradesh','arunachal-pradesh','assam','bihar',
                              'chandigarh','chhattisgarh','dadra-&-nagar-haveli-&-daman-&-diu','delhi','goa','gujarat','haryana',
                              'himachal-pradesh','jammu-&-kashmir','jharkhand','karnataka','kerala','ladakh','lakshadweep',
                              'madhya-pradesh','maharashtra','manipur','meghalaya','mizoram',
                              'nagaland','odisha','puducherry','punjab','rajasthan','sikkim',
                              'tamil-nadu','telangana','tripura','uttar-pradesh','uttarakhand','west-bengal'),index=30)
            
            query = f"SELECT Pincode, sum(Transaction_count) as Total_Transactions, sum(Transaction_amount) as Total_amount from top_trans_pin WHERE Year BETWEEN 2018 AND 2022 AND Quarter BETWEEN 1 AND 4  AND State = '{selected_state5}' GROUP by State,Pincode,Year,Quarter ORDER by State,Pincode"
            cur.execute(query)
            results = cur.fetchall()
            df_results = pd.DataFrame(results, columns=['Pincode', 'Total_Transactions','Total_amount'])
            df_results['var_amount'] = np.log10(df_results['Total_amount'])
            fig5 = px.pie(df_results, values='var_amount', names='Pincode', hover_data='Total_Transactions',
                  title=  selected_state5)
            col1, col2=st.columns(2)
            col1.plotly_chart(fig5,use_container_width=True)
            col2.markdown('**Top Transactions by Pincode**')
            col2.write(df_results)
            
            
            
    with tab6:  
            selected_state6 = st.selectbox("STATES6",
                             ('andaman-&-nicobar-islands','andhra-pradesh','arunachal-pradesh','assam','bihar',
                              'chandigarh','chhattisgarh','dadra-&-nagar-haveli-&-daman-&-diu','delhi','goa','gujarat','haryana',
                              'himachal-pradesh','jammu-&-kashmir','jharkhand','karnataka','kerala','ladakh','lakshadweep',
                              'madhya-pradesh','maharashtra','manipur','meghalaya','mizoram',
                              'nagaland','odisha','puducherry','punjab','rajasthan','sikkim',
                              'tamil-nadu','telangana','tripura','uttar-pradesh','uttarakhand','west-bengal'),index=30)
            
            
            query = f"SELECT Pincode, sum(RegisteredUsers) as Total_Users FROM  top_user_pin WHERE Year BETWEEN 2018 AND 2022 AND Quarter BETWEEN 1 AND 4  AND State = '{selected_state6}' GROUP by State,Pincode,Year,Quarter order by State,Pincode"
            cur.execute(query)
            results = cur.fetchall()
            df_results = pd.DataFrame(results, columns=['Pincode', 'Total_Users '])
            
            fig6 = px.pie(df_results, values='Total_Users ', names='Pincode', hover_data='Total_Users ',
                  title=  selected_state6)
            col1, col2=st.columns(2)
            col1.plotly_chart(fig6,use_container_width=True)
            col2.markdown('**Top Users by Pincode**')
            col2.write(df_results)
            
            
            
            
if selected == "Map_Visualizations":            
    tab1,tab2= st.tabs(['Map of Statewise Transactions' , ' Map of Statewise Users'])       
    
    with tab1:
            query = f"SELECT State, sum(Transaction_count) as Total_Transactions, sum(Transaction_amount) as Total_amount FROM  map_trans WHERE Year BETWEEN 2018 AND 2022 AND Quarter BETWEEN 1 AND 4 GROUP BY State, Year, Quarter ORDER BY State, Year, Quarter"    
            cur.execute(query)
            results = cur.fetchall()
            df_results = pd.DataFrame(results, columns=['Total_amount', 'Total_Transactions', 'State'])
            df_results.Total_Transactions= df_results.Total_Transactions.astype(float)
            #df_results['var_amount'] = np.log10(df_results['Total_Transactions'])
    
    
            df2 = pd.read_csv('C:/Users/ADMIN/Documents/state_names.csv')
            df_results['State'] = df2['State Names']
   
            fig = px.choropleth(df_results,geojson=geojson_data,
                      featureidkey='properties.st_nm',
                      locations='State',
                      scope='asia',
                      color='Total_Transactions',
                      hover_name='State',
                      title='Map Transactions Statewise On The Basis of Amount',  
                      color_continuous_scale='ylorrd')

            fig.update_geos(fitbounds="locations", visible=False)
            fig.show()
            st.write(df_results)
            
            
    with tab2:  
            query = f"SELECT State, sum(RegisteredUser) as Total_Users, sum(AppOpens) as Total_AppOpens FROM  map_user WHERE Year BETWEEN 2018 AND 2022 AND Quarter BETWEEN 1 AND 4 GROUP BY State ORDER BY State"    
            cur.execute(query)
            results = cur.fetchall()
            df_results = pd.DataFrame(results, columns=['State', 'Total_Users', 'Total_AppOpens'])
            df_results.Total_AppOpens = df_results.Total_AppOpens.astype(float)
            
            df2 = pd.read_csv('C:/Users/ADMIN/Documents/state_names.csv')
            df_results['State'] = df2['State Names']
   
            fig1 = px.choropleth(df_results,geojson=geojson_data,
                      featureidkey='properties.st_nm',
                      locations='State',
                      scope='asia',
                      color='Total_AppOpens',
                      #hover_name='State',
                      title='Map Users Statewise On The Basis of Users',  
                      color_continuous_scale='ylorrd')

            fig1.update_geos(fitbounds="locations", visible=False)
            fig1.show()
            st.write(df_results)
```
